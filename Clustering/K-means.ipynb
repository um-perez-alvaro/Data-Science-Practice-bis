```python
import numpy as np
import matplotlib.pyplot as plt
```

# K-means algorithm

## Example

```python
blob_centers = np.array(
        [[0.2, 2.3],
         [-1.5,2.3],
         [-2.8,1.8],
         [-2.8,2.8],
         [-2.8,1.3]]
)
blob_std = np.array([0.4,0.3,0.1,0.1,0.1])
```

```python
from sklearn.datasets import make_blobs
X,y = make_blobs(n_samples=2000, centers = blob_centers, cluster_std=blob_std)
```

```python
plt.figure(figsize=(15,8))
plt.plot(X[:,0],X[:,1],'o',markersize=3)
```

Let us train a k-means clusterer on this dataset.

```python
from sklearn.cluster import KMeans
```

```python
k = 5 # number of cluster 
kmeans = KMeans(n_clusters=k)
kmeans.fit(X)
y = kmeans.predict(X)
```

Each instance was assigned to one of the 5 cluster

```python
y
```

And the following 5 cluster centers (means) were estimated

```python
centers = kmeans.cluster_centers_
centers
```

```python
plt.figure(figsize=(15,8))
colors = ['blue','red','green','yellow','orange']
for i in range(k):
    color = colors[i]
    plt.plot(X[y==i,0],X[y==i,1],'o',color=color,markersize=3, label=str(i)+'th cluster') # ith cluster
    plt.plot(centers[i,0],centers[i,1],'x',color='black',markersize=20,markeredgewidth=5)# ith center
plt.legend()
```

## How does (the original) kmeans work?

- First initialize the cluster centers randomly (chosen randomly from the dataset)
- Repeat until cluster centers do not move:
    - Assign each instance to the closest cluster center
    - Update the cluster centers to be the mean of the instances that are assigned to them

Let us run the k-means algorithm for 1, 2 and 3 iterations (to see how the cluster centers move around)

```python
kmeans_iter1 = KMeans(n_clusters=5,
                      n_init=1, # Number of time the k-means algorithm will be run 
                      algorithm='full', 
                      init='random', # original kmeans
                      max_iter=1, 
                      random_state=1) # always the same random initialization
kmeans_iter2 = KMeans(n_clusters=5, n_init=1, algorithm='full', init='random', max_iter=2, random_state=1)
kmeans_iter3 = KMeans(n_clusters=5, n_init=1, algorithm='full', init='random', max_iter=3, random_state=1)
kmeans_iter4 = KMeans(n_clusters=5, n_init=1, algorithm='full', init='random', max_iter=4, random_state=1)
kmeans_iter5 = KMeans(n_clusters=5, n_init=1, algorithm='full', init='random', max_iter=5, random_state=1)
kmeans_iter6 = KMeans(n_clusters=5, n_init=1, algorithm='full', init='random', max_iter=6, random_state=1)
```

```python
kmeans_iter1.fit(X)
kmeans_iter2.fit(X)
kmeans_iter3.fit(X)
kmeans_iter4.fit(X)
kmeans_iter5.fit(X)
kmeans_iter6.fit(X)
```

```python
y = kmeans_iter1.predict(X)
y
```

```python
# initialization
y = kmeans_iter1.predict(X)
centers = kmeans_iter1.cluster_centers_
plt.figure(figsize=(15,8))
colors = ['blue','red','green','yellow','orange']
for i in range(k):
    color = colors[i]
    plt.plot(X[y==i,0],X[y==i,1],'o',color=color,markersize=3, label=str(i)+'th cluster') # ith cluster
    plt.plot(centers[i,0],centers[i,1],'x',color='black',markersize=20,markeredgewidth=5)# ith center
plt.legend()
```

```python
# iter 2
y = kmeans_iter2.predict(X)
centers = kmeans_iter2.cluster_centers_
plt.figure(figsize=(15,8))
colors = ['blue','red','green','yellow','orange']
for i in range(k):
    color = colors[i]
    plt.plot(X[y==i,0],X[y==i,1],'o',color=color,markersize=3, label=str(i)+'th cluster') # ith cluster
    plt.plot(centers[i,0],centers[i,1],'x',color='black',markersize=20,markeredgewidth=5)# ith center
plt.legend()
```

```python
# iter 3
y = kmeans_iter3.predict(X)
centers = kmeans_iter3.cluster_centers_
plt.figure(figsize=(15,8))
colors = ['blue','red','green','yellow','orange']
for i in range(k):
    color = colors[i]
    plt.plot(X[y==i,0],X[y==i,1],'o',color=color,markersize=3, label=str(i)+'th cluster') # ith cluster
    plt.plot(centers[i,0],centers[i,1],'x',color='black',markersize=20,markeredgewidth=5)# ith center
plt.legend()
```

```python
# iter 6
y = kmeans_iter6.predict(X)
centers = kmeans_iter6.cluster_centers_
plt.figure(figsize=(15,8))
colors = ['blue','red','green','yellow','orange']
for i in range(k):
    color = colors[i]
    plt.plot(X[y==i,0],X[y==i,1],'o',color=color,markersize=3, label=str(i)+'th cluster') # ith cluster
    plt.plot(centers[i,0],centers[i,1],'x',color='black',markersize=20,markeredgewidth=5)# ith center
plt.legend()
```
