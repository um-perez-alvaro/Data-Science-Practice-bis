```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Subplots, figures and axes

All plotting is done with respect to an Axes. Typically, you'll set up a Figure, and then add an Axes to it. You can use fig.add_axes, but in most cases, fig.add_subplot will fits your needs perfectly.

```python
fig, axes = plt.subplots(nrows=2,ncols=3, figsize=(15,10)) # figsize = (width, heigh) 
plt.show
```

When plt.subplots() is called with no arguments, it creates a new figure with a single subplot.

```python
fig, axes = plt.subplots()
plt.show
```

```python
# generate some random data
x = np.random.randn(20) # 20 random points

# plot the data
fig, axes = plt.subplots()
axes.plot(x)
```

## Colors, Markers and Line Styles

[pyplot.plot documentation](https://matplotlib.org/3.2.2/api/_as_gen/matplotlib.pyplot.plot.html)

```python
fig, axes = plt.subplots(nrows=2,ncols=3,figsize=(15,10))
# line plot
axes[0,0].plot(x) 
# line + markers
axes[0,1].plot(x,marker='o') # abbreviation: plot(x,'o-')
# dashed line + markers 
axes[0,2].plot(x,marker='o',linestyle='--') # abbreviation: plot(x,'o--')
# markers
axes[1,0].plot(x,marker='o',linestyle='None') # abbreviation: plot(x,'o')
# dashed line + markers, in red
axes[1,1].plot(x,marker='o',linestyle='--', color='r')  # abbreviation: plot(x,'ro--')
# dashed red line + blue markers
axes[1,2].plot(x,marker='o',markerfacecolor='b',markeredgecolor='b',linestyle='--',color='r')
```

## Ticks, Labels, and Legends

```python
x = np.random.randn(1000).cumsum()
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(x)

# change x-axis ticks
ax.set_xticks([0,50,100,200,400,800])
ax.set_xticklabels(['zero','fifty','a hundred','two hundreds','four hundreds','eight hundreds'], rotation=90,
                  fontsize=15)
ax.set_xlabel('x-axis',fontsize=20)
# change y-axis ticks
ax.set_ylabel('y-axis',fontsize=20)
ax.set_yticks([10,0,-10,-20,-30])
# change y-axis ticks size without changing the labels
ax.tick_params(axis='y', labelsize=15)
# set title
ax.set_title('A matplotlib plot',fontsize=30)
plt.show()
```

```python
x1 = np.random.randn(1000).cumsum()
x2 = np.random.randn(1000).cumsum()
x3 = np.random.randn(1000).cumsum()
```

```python
fig, ax = plt.subplots(figsize=(12,5))
ax.plot(x1, color='blue', label='one')
ax.plot(x2, color='red', label='two')
ax.plot(x3, color='green', label='three')

ax.legend(fontsize=15, loc='best')
```

## Plotting with pandas

```python
# load the data
url = 'https://raw.githubusercontent.com/um-perez-alvaro/Data-Science-Practice/master/Pandas/Data/Advertising.csv'
advertising = pd.read_csv(url)
advertising.drop('Unnamed: 0',axis=1,inplace=True)
advertising.head()
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2, figsize=(15,10))

advertising.plot.scatter(x='TV',y='Sales',ax=axes[0,0],c='r')
advertising.plot.scatter(x='Radio',y='Sales',ax=axes[0,1],c='g')
advertising.plot.scatter(x='Newspaper',y='Sales',ax=axes[1,0],c='b')
advertising.plot.scatter(x='TV',y='Radio',ax=axes[1,1],c='m')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(15,15))

advertising.loc[0:10,['TV','Radio','Newspaper']].plot(kind='bar', ax=axes[0])
advertising.loc[0:10,['TV','Radio','Newspaper']].plot(kind='bar',stacked=True, ax=axes[1])
```

```python
fig, ax = plt.subplots(figsize=(12,7))

plot1 = advertising.loc[0:10,['TV','Radio','Newspaper']].plot(kind='bar',stacked=True, ax=ax)
ax.set_ylabel('Budget')
ax.legend(loc='upper center')
# secondary axis
ax2 = ax.twinx()  # instantiate a second axes that shares the same x-axis
ax2.set_ylabel('Sales')
plot2 = advertising.Sales[0:11].plot(ax=ax2,marker='o', color='red',linestyle='--', label='Sales')
ax2.legend(loc='upper right')
```

```python
fig, axes = plt.subplots(nrows=2,ncols=1, figsize=(15,15))

advertising.Sales.plot.hist(ax=axes[0], bins=20)
advertising.Sales.plot.density(ax=axes[1])
```

```python
fig, ax = plt.subplots(figsize=(15,7))
advertising.loc[0:10,['Radio','Newspaper','TV']].plot(marker='o', linestyle='--',ax=ax)

ax.set_title('TV, Radio and Newspaper budgets', fontsize=20)
plt.legend(fontsize=20,loc='best')
```

```python
fig, ax = plt.subplots(figsize=(15,7))
advertising.loc[0:10,['Radio','Newspaper','TV']].plot.area(ax=ax,stacked=True)

ax.set_title('TV, Radio and Newspaper budgets', fontsize=20)
plt.legend(fontsize=20)
```
