```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Efficiently searching for optimal tuning parameters: GridSearchCV

```python
# load the data the iris dataset
url = 'https://raw.githubusercontent.com/um-perez-alvaro/lin-regress/master/iris.data'
iris = pd.read_csv(url, header=None, names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width','species'])
X = iris.iloc[:,0:4] # feature matrix
y = iris.species
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier(n_neighbors=5)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# define the parameter values that should be searched
k_range = list(range(1, 31))
print(k_range)
```

```python
# create a parameter grid: map the parameter names to the values that should be searched
param_grid = dict(n_neighbors=k_range)
print(param_grid)
```

```python
# instantiate the grid
grid = GridSearchCV(knn_clf, param_grid, cv=10, scoring='accuracy', return_train_score=False)
```

```python
grid.fit(X,y)
```

```python
pd.DataFrame(grid.cv_results_)[['mean_test_score', 'std_test_score', 'params']]
```

## Searching multiple parameters simultaneously

```python
# define the parameter values that should be searched
k_rankge = list(range(1,31))
weight_options = ['uniform','distance']
```

```python
# create a parameter grid
param_grid = dict(n_neighbors = k_range, weights=weight_options)
print(param_grid)
```

```python
# instantiate and fit the grid
grid = GridSearchCV(knn_clf, param_grid, cv=10, scoring='accuracy', return_train_score=False)
grid.fit(X, y)
```

```python
# view the results
pd.DataFrame(grid.cv_results_)[['mean_test_score', 'std_test_score', 'params']]
```

```python
# examine the best model
print(grid.best_score_)
print(grid.best_params_)
```
