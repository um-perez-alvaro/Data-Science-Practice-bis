# Adjusting the classification threshold

## Loading the data

from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784',version=1)

```python
X, y = mnist['data'],mnist['target']
```

There are 70,000 images, and each image has 784 features. This is because each image is 28×28 pixels, and each feature simply represents one pixel's intensity, from 0 (white) to 255 (black).

```python
X.shape
```

Let's take a peek at one digit from the dataset. All we need to do is grab an instance's feature vector, reshape it to a 28×28 array, and display it using imshow()

```python
some_digit = X[1]
some_digit_image = some_digit.reshape(28,28)
plt.imshow(some_digit_image, cmap = 'binary')
plt.axis('off')
```

This looks like a 5, and indeed that's what the label tells us

```python
y[1]
```

The following figure shows a few more images from the MNIST dataset

```python
# display the first 100 digits
plt.figure(figsize=(10,10))
for i in range(100):
    some_digit = X[i] #ith digit
    some_digit_image = some_digit.reshape(28,28) # reshape it into a 28x28 matrix
    plt.subplot(10,10,i+1) # select subplot
    plt.imshow(some_digit_image, cmap = 'binary') # plot digit image
    plt.axis('off')
```

**Goal:** to train a 5-detector

```python
'create a dummy variable (1 if digit is 5, 0 otherwise)'
y_5 = np.zeros(len(y))
y_5[y=='5'] = 1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_5, test_size=0.3)
```

```python
from sklearn.linear_model import SGDClassifier
```

```python
sgd = SGDClassifier() # create the classifier
sgd.fit(X_train/255,y_train)
```

## Evaluation Metrics

```python
# make class predictions for the testing set
y_test_pred = sgd.predict(X_test/255)
```

```python
print(100*metrics.precision_score(y_test, y_test_pred))
```

```python
print(100*metrics.recall_score(y_test, y_test_pred))
```

## Precision/Recall Trade-off

```python
from sklearn import metrics
```

```python
y_test_scores = sgd.decision_function(X_test/255)
```

```python
precisions, recalls, thresholds = metrics.precision_recall_curve(y_test, y_test_scores)
```

```python
plt.figure(figsize=(12,5))
plt.plot(thresholds,precisions[:-1],'b--',label='Precision')
plt.plot(thresholds,recalls[:-1],'g-', label='Recall')
plt.legend(loc="center right", fontsize=16) 
plt.xlabel("Threshold", fontsize=16)        
plt.grid(True)                             
plt.axis([thresholds.min(),thresholds.max(),0,1.1])  
plt.title('Precision and Recall Curves',fontsize=20)
```

```python
plt.figure(figsize=(12,5))
plt.plot(recalls, precisions, "b-", linewidth=2)
plt.xlabel("Recall", fontsize=16)
plt.ylabel("Precision", fontsize=16)
plt.axis([0, 1, 0, 1])
plt.grid(True)
plt.title('Precision vs Recall Curve',fontsize=20)
```

```python
fpr,tpr, thresholds = metrics.roc_curve(y_test, y_test_scores)
```

```python
plt.figure(figsize=(12,5))
plt.plot(fpr, tpr, linewidth=2)
plt.plot([0, 1], [0, 1], 'k--') # dashed diagonal
plt.axis([0, 1, 0, 1])                                    
plt.xlabel('False Positive Rate (Fall-Out)', fontsize=16) 
plt.ylabel('True Positive Rate (Recall)', fontsize=16)    
plt.grid(True)          
plt.title('ROC Curve',fontsize=20)
```

```python
metrics.roc_auc_score(y_test, y_test_scores)
```

## Pima Dataset

### logistic regression

```python
# Pima indian diabetes dataset
pima = pd.read_csv('Data/pima.csv')
```

```python
# define X and y
feature_cols = ['pregnant','glucose','bp','skin','insulin','bmi','age']
X = pima[feature_cols]
y = pima.label
```

```python
# split X and y into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
y_scores = cross_val_predict(logreg,X_train,y_train, cv=5, method='decision_function')
```

```python
from sklearn.metrics import precision_recall_curve
precisions, recalls, thresholds = precision_recall_curve(y_train,y_scores)
```

```python
plt.figure(figsize=(12,5))
plt.plot(thresholds,precisions[:-1],'b--',label='Precision')
plt.plot(thresholds,recalls[:-1],'g-', label='Recall')
plt.legend(loc="center right", fontsize=16) 
plt.xlabel("Threshold", fontsize=16)        
plt.grid(True)                             
plt.axis([thresholds.min(),thresholds.max(),0,1.1])  
plt.title('Precision and Recall Curves',fontsize=20)
```

```python
plt.figure(figsize=(12,5))
plt.plot(recalls, precisions, "b-", linewidth=2)
plt.xlabel("Recall", fontsize=16)
plt.ylabel("Precision", fontsize=16)
plt.axis([0, 1, 0, 1])
plt.grid(True)
plt.title('Precision vs Recall Curve',fontsize=20)
```

```python
from sklearn.metrics import roc_curve
fpr,tpr,thresholds = roc_curve(y_train,y_scores)
```

```python
plt.figure(figsize=(12,5))
plt.plot(fpr, tpr, linewidth=2)
plt.plot([0, 1], [0, 1], 'k--') # dashed diagonal
plt.axis([0, 1, 0, 1])                                    
plt.xlabel('False Positive Rate (Fall-Out)', fontsize=16) 
plt.ylabel('True Positive Rate (Recall)', fontsize=16)    
plt.grid(True)          
plt.title('ROC Curve',fontsize=20)
```
