```python
import pandas as pd
```

```python
good_or_evil_train = pd.DataFrame({'name': ['batman','robin','alfred','penguin','catwoman','joker'], 
                                   'sex' : ['male','male','male','male','female','male'],
                                   'mask' : ['yes','yes','no','no','yes','no'],
                                   'cape' : ['yes','yes','no','no','no','no'],
                                   'tie' : ['no','no','yes','yes','no','no'],
                                   'ears' : ['yes','no','no','no','yes','no'],
                                   'smokes' : ['no','no','no','yes','no','no'],
                                   'class' : ['good','good','good','bad','bad','bad']}).set_index('name')
good_or_evil_test = pd.DataFrame({'name': ['batgirl','riddler'], 
                                   'sex' : ['female','male'],
                                   'mask' : ['yes', 'yes'],
                                   'cape' : ['yes','no'],
                                   'tie' : ['no','no'],
                                   'ears' : ['yes','no'],
                                   'smokes' : ['no','no']}).set_index('name')
```
